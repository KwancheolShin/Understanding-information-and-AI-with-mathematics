#1변수 함수의 경사하강법

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def f(x):
    return x**2 - 4*x + 5

# 도함수 정의
def grad_f(x):
    return 2*x - 4

# 수치 미분계수 정의
def grad_fh(x,h = 1e-5):
    return (f(x+h)-f(x))/h

# 하이퍼파라미터 셋팅
lr = 0.1    # 학습률 # 1보다 크면 발산, 0.5에서 한방에
epochs = 1000 # 반복 횟수
num_pt = 20   # 출력 갯수

# 배열 x_n 초기화
x_n = np.zeros(epochs + 1) # 수열 x_n 이 들어갈 자리 초기화
x_n[0] = 4.0 # x_0 값 입력

# 경사 하강법 루프
for i in range(epochs):
    x_n[i+1] = x_n[i] - lr * grad_f(x_n[i])
    if i < num_pt:
        print(f"Step {i+1:3d}: x = {x_n[i+1]:.6f}, f(x) = {f(x_n[i+1]):.6f}")

# 함수 그래프용 x 값과 f(x)
x_range = np.linspace(0, 5, 100) # 정의역을 0부터 5까지, 100개의 점을 일정한 간격으로 생성
f_range = f(x_range)             # 치역. f에 넘파이 배열을 입력해도 알아서 각각 계산 => 넘파이의 장점
f_vals  = f(x_n)              # (x_n, f(x_n))

# 시각화
plt.figure(figsize=(8, 5))
plt.plot(x_range, f_range, label='Loss Function $f(x)$') #(x,f(x))
plt.plot(x_n[:num_pt+1], f_vals[:num_pt+1], 'b.', label="Gradient Steps") #(x_n,f(x_n))

# 화살표로 연결하는 방법 (이런게 있다 정도만...)
for i in range(num_pt):
    plt.annotate(
        '',
        xy=(x_n[i+1], f_vals[i+1]),
        xytext=(x_n[i], f_vals[i]),
        arrowprops=dict(arrowstyle='->', color='red', lw=1),
    )

plt.xlabel("x")    # x 축 이름
plt.ylabel("f(x)") # y 축 이름
plt.title("Gradient Descent with Arrows")
plt.legend()   # 범례 생성
plt.grid(True) # 격자무늬 생성
plt.show()     # 그림 짜잔~
```
