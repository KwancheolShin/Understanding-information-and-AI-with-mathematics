```python
import numpy as np
import matplotlib.pyplot as plt


ff = lambda x, y: 3*(x-6)**2 + (y-5)**2
fx = lambda x, y: 6*(x-6)
fy = lambda x, y: 2*(y-5)

ox, oy = 1, 1 # 
epochs = 100
lr = 0.05 #0.25 0.3

xp, yp = np.zeros(epochs+1), np.zeros(epochs+1)

xp[0], yp[0] = ox, oy

for i in range(epochs):
    dx, dy = fx(ox,oy), fy(ox,oy)   # Compute gradient
    nx, ny = ox - lr*dx, oy - lr*dy # Gradient Descent
    ox, oy = nx, ny                 # Update Iteration
    xp[i+1], yp[i+1] = ox, oy

l_bdr, r_bdr = 0, 10
x = np.linspace(l_bdr,r_bdr)
y = np.linspace(l_bdr,r_bdr)

xx, yy = np.meshgrid(x,y) #  
zz = ff(xx,yy)

#  
fig1 = plt.figure()
ax1  = plt.axes()
ax1.contour(xx, yy, zz, 20, cmap='jet')
ax1.set_aspect('equal')
ax1.plot(6, 5, 'rs')
ax1.plot(xp, yp, 'ko-')
fig1.show()  #  plt.show()   

# 3 
fig2 = plt.figure()
ax2 = plt.axes(projection='3d')
ax2.plot_surface(xx, yy, zz, cmap='jet')
ax2.scatter(6, 5, ff(6, 5), color='r', s=50)
ax2.plot(xp, yp, ff(xp, yp), 'k.')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x, y)')
fig2.show()
```
