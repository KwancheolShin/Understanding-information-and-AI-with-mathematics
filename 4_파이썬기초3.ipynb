#14. 넘파이
넘파이의 배열 클래스인 numpy.array에는 편리한 메서드가 많이 준비되어 있어 배열이나 행렬 계산이 많이 등장하는 딥러닝을 구현할 때 이 메서드들을 이용한다.


```python
import numpy as np #numpy 라이브러리 가져오기
x = np.array([1.0,2.0,3.0])
print("x=", x)
print("type=", type(x))
print("shape = ",x.shape)
print("size = ", x.size)
print("dimension of x = ", x.ndim)
```

##14.1. 넘파이 산술연산

```python
y = np.array([2.0,4.0,6.0])

print("x+y = ", x+y)
print("x-y = ", x-y)
print("x/y = ", x-y)
print("x*y = ", x*y)
```

##14.2. 넘파이의 N차원 배열
넘파이는 1차원 배열(1줄로 늘어선 배열)뿐만 아니라 다차원 배열도 작성할 수 있다. 1차원 배열, 2차원 배열처럼 원하는 차수의 배열을 만들 수 있다. 이러한 N차원 배열을 수학에서는 텐서tensor라고도 한다.

```python
A = np.array([[1,2],[3,4]])
B = np.array([[3,0],[0,6]])
print(A)
print(A+B)
print(A*B) #같은 위치에 성분끼리 곱해진다. "Hadamard곱"이라 한다.
```

```python
print(A.shape) #shape
print(A.size)  #총 원소의 갯수
print(A.ndim)  #dimension을 보여줌
print(A.dtype) #data type을 보여줌
```

#15. matplotlib
딥러닝 실험에서는 그래프 그리기와 데이터 시각화가 중요하다. **matplotlib**는 그래프를 그려주는 라이브러리이다. matplotlib를 사용하면 그래프 그리기와 데이터 시각화가 쉬워진다.


##단순한 그래프 그리기
그래프를 그리려면 matplotlib의 **pyplot** 모듈을 이용한다.

```python
import numpy as np
import matplotlib.pyplot as plt

#데이터 준비
x = np.arange(0,6,0.1) # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x,y1, label='sin')
plt.plot(x,y2, linestyle = '--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```

#16. input()함수로 사용자의 입력을 받기
기본적으로 입력값은 문자열(str) 형태로 반환된다.

```python
name = input("이름을 입력하세요: ")
#실행하면 "이름을 입력하세요: "라는 프롬프트가 나타나고, 사용자가 입력한 값을 name 변수에 저장한다.
print(f"제 이름은 {name} 입니다.")
```

```python
#숫자 입력 받기 (int, float 변환)
age = int(input("나이를 입력하세요: "))
print("당신의 나이는", age, "살입니다.")
```

```python
#여러 개의 입력값 받기 (split() 사용)
a, b = input("두 개의 숫자를 입력하세요: ").split()
#split()은 공백을 기준으로 입력을 나눕니다.
a, b = int(a), int(b)  # 문자열을 정수로 변환
print("두 숫자의 합:", a + b)
```
