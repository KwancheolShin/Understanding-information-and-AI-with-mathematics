```python
# -*- coding: utf-8 -*-
"""
Created on Sun Apr 20 04:59:21 2025

@author: kcshi
"""
import numpy as np
import matplotlib.pyplot as plt


#def step_func(x):
#    return np.array(x>0, dtype = int)
# def logistic(x):
#     return 1/(1+np.exp(-x))
# def tanh(x):
#     return np.tanh(x)
# def arctan(x):
#     return np.arctan(x)
# def relu(x):
#     return np.maximum(0,x)


step_func = lambda x: np.array(x>0, dtype = int)
logistic  = lambda x :1/(1+np.exp(-x))
tanh      = lambda x: np.tanh(x)
arctan    = lambda x: np.arctan(x)
relu      = lambda x: np.maximum(0,x)

dlogistic = lambda x: logistic(x)*(1-logistic(x))
dtanh     = lambda x: 1 - tanh(x)**2
darctan   = lambda x: np.cos(arctan(x))**2

x = np.linspace(-6, 6, 400)

plt.figure(figsize=(10, 6))
plt.plot(x, step_func(x), 'k--')
plt.title('step function')
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid()
plt.show()


# plt.figure(figsize=(10, 6))
# plt.plot(x, step_func(x), 'k--')
# plt.plot(x,logistic(x),'k-')
# plt.title('step vs logistic')
# plt.xlabel("x")
# plt.ylabel("h(x)")
# plt.legend(('step', 'logistic'))
# plt.grid()
# plt.show()


plt.figure(figsize=(10, 6))
plt.plot(x, logistic(x),  'k-')
plt.plot(x, dlogistic(x), 'b-')
plt.title("logistic(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(('h(x)','diff h(x)'))
plt.grid()
#plt.grid(True)
#plt.axis('square')
#plt.axhline(y=0, color='black', label='x')
#plt.axvline(x=0, color='black', label='y')
plt.show()



plt.figure(figsize=(10, 6))
plt.plot(x, tanh(x), 'k-')
plt.plot(x, dtanh(x), 'b-')
plt.title("tanh(x)")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.legend(('h(x)','diff h(x)'))
plt.grid()
#plt.grid(True)
#plt.axis('square')
#plt.axhline(y=0, color='black', label='x')
#plt.axvline(x=0, color='black', label='y')
plt.show()



plt.figure(figsize=(10, 6))
plt.plot(x, arctan(x), 'k-')
plt.plot(x, darctan(x),'b-')
plt.title("arctan(x)")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.legend(('h(x)','diff h(x)'))
plt.grid()

#plt.grid(True)
#plt.axis('square')
#plt.axhline(y=0, color='black', label='x')
#plt.axvline(x=0, color='black', label='y')
plt.show()



plt.figure(figsize=(10, 6))
plt.plot(x, relu(x), 'k-')
plt.plot(x,step_func(x), 'b-')
plt.title("ReLU(x)")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.legend(('h(x)','diff h(x)'))
plt.grid()
#plt.grid(True)
#plt.axis('square')
#plt.axhline(y=0, color='black', label='x')
#plt.axvline(x=0, color='black', label='y')
plt.show()
```
