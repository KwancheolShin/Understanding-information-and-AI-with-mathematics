```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegressionGD:
    def __init__(self, learning_rate=0.001, epochs=20000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.w = 0.0
        self.b = 0.0
        self.cost_history = []

    def compute_cost(self, x, y):
        y_pred = self.w * x + self.b
        cost = np.mean((y - y_pred) ** 2)
        return cost

    def fit(self, x, y, verbose=True):
        n = len(x)

        for epoch in range(self.epochs):
            y_pred = self.w * x + self.b
            error = y_pred - y

            # Gradient calculation
            dw = (2 / n) * np.sum(error * x)
            db = (2 / n) * np.sum(error)

            # Update parameters
            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

            # Cost logging
            cost = self.compute_cost(x, y)
            self.cost_history.append(cost)

            if verbose and epoch % 10 == 0:
                print(f"Epoch {epoch}: Cost={cost:.4f}, w={self.w:.4f}, b={self.b:.4f}")

    def predict(self, x):
        return self.w * x + self.b

    def plot_fit_and_cost(self, x, y):
        y_fit = self.predict(x)

        plt.figure(figsize=(10, 4))

        # 1. Regression result
        plt.subplot(1, 2, 1)
        plt.scatter(x, y, label='Data')
        plt.plot(x, y_fit, color='red', label='Regression line')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Linear Regression Fit')
        plt.legend()
        plt.grid(True)

        # 2. Cost history
        plt.subplot(1, 2, 2)
        plt.plot(self.cost_history, label='Cost (MSE)')
        plt.xlabel('Epoch')
        plt.ylabel('Cost')
        plt.title('Cost History')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()
```

```python
# 데이터
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])

# 모델 생성 및 학습
model = LinearRegressionGD(learning_rate=0.0001, epochs=20000)
model.fit(x,y)
```

```python
# 예측
print("\n최종 파라미터:")
print(f"w = {model.w:.4f}, b = {model.b:.4f}")

# 시각화
model.plot_fit_and_cost(x, y)
```

```python
print("predicted value for input 10 is", model.predict(10))
```
